## Try and Except Block to filter out the Corrupted Images

```python
data = []

for category in CATEGORIES:
  path = os.path.join(DIRECTORY , category)                                   
  for img in os.listdir(path):#labelling the images in directory according to categorical names
    try:
        img_path = os.path.join(path,img)#getting the image path
        label = CATEGORIES.index(category)##labelling Covid19 Negative as 0 and Covid19 Positive as 1
        arr = cv2.imread(img_path, cv2.IMREAD_GRAYSCALE) #converting all the images to Grayscale for faster computation
        new_arr = cv2.resize(arr, (img_size, img_size)) #resizing all the images to 100 x 100 matrix size
        data.append([new_arr, label])#appending image and label in list
    except Exception as e:
        print(str(e))

```

## Seperating Features and Target Variables

```python
X = []
y = []
```

```python
for features,label in data:
    X.append(features) #Storing all Images in X
    y.append(label)  #Storing all Labels in y
```

## The two lists X and y are converted to Numpy arrays for Tensorflow understanding

```python
X = np.array(X)
y = np.array(y)
```
